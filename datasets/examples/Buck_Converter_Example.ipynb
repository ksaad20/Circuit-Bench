# Buck Converter Example

CircuitBench synthetic buck converter dataset.

```python
import numpy as np
import pandas as pd

Vin = np.linspace(12, 24, 20)
Duty = np.linspace(0.2, 0.8, 20)

records = []

for vin in Vin:
    for duty in Duty:
        vout = vin * duty
        efficiency = 0.92 - 0.05 * (1 - duty) - 0.001 * (vin - 12)

        records.append(
            {
                "Input_Voltage": vin,
                "Duty_Cycle": duty,
                "Output_Voltage": vout,
                "Efficiency": efficiency,
                "Topology": "Buck_Converter",
            }
        )

df = pd.DataFrame(records)

df.to_csv("buck_converter_dataset.csv", index=False)

df.head()
```
