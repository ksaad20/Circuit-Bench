# Analog Filters Example

CircuitBench synthetic RC low-pass filter dataset.

```python
import numpy as np
import pandas as pd

R_values = np.array([100,220,470,1000])
C_values = np.array([1e-6,2.2e-6,4.7e-6])

records=[]

for R in R_values:
    for C in C_values:
        cutoff = 1/(2*np.pi*R*C)
        records.append({
            "Resistance_Ohm": R,
            "Capacitance_F": C,
            "Cutoff_Hz": cutoff,
            "Filter_Type": "RC_Low_Pass"
        })

df = pd.DataFrame(records)

df.to_csv(
    "analog_filter_dataset.csv",
    index=False
)

df.head()
```
